```python
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit

# Create a dummy dataset
import pandas as pd
df = pd.DataFrame({'feature_1': [1, 2, 3, 4, 5],
                   'feature_2': ['A', 'B', 'B', 'C', 'C'],
                   'target': [0, 1, 1, 0, 0]})

# Define the stratified sampler
stratified_sampler = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

# Split the dataset using the stratified sampler
for train_index, test_index in stratified_sampler.split(df, df['target']):
    train_set = df.loc[train_index]
    test_set = df.loc[test_index]
    
# Check the number of samples in each set
print(train_set['target'].value_counts())
print(test_set['target'].value_counts())
```

# 2.Quota Sampling



###          :                       .   ,    

```python
# no specific library in Python
import numpy as np

# Assuming we have a dataset 'data' with two variables, 'variable1' and 'variable2',
# and we want to perform quota sampling with a sample size of 100 and a quota of 40% for 'variable1'
sample_size = 100
quota_variable1 = 0.4

# Compute the number of observations to select for each level of 'variable1'
n_variable1 = int(sample_size * quota_variable1)
n_variable2 = sample_size - n_variable1

# Select the first n_variable1 observations from 'data' that meet the quota for 'variable1'
sample_variable1 = data[data['variable1'] == some_value][:n_variable1]

# Select the remaining n_variable2 observations from 'data' that meet the quota for 'variable2'
sample_variable2 = data[data['variable2'] == some_other_value][:n_variable2]

# Combine the two samples to form the final quota sample
sample = pd.concat([sample_variable1, sample_variable2])
```

# 4. Snowball Sampling










## Like Nokia Snake game

```python
# no specific library in Python
import numpy as np

# Assuming we have a dataset 'data' and we want to perform snowball sampling with a maximum depth of 3 and a starting observation
start_observation = 10
max_depth = 3

# Initialize the sample with the starting observation
sample = [start_observation]

# Perform snowball sampling by iteratively selecting observations that are referred to by the current sample
for i in range(max_depth):
    current_sample = sample[-1]
    referred_observations = data[data['referral_variable'] == current_sample]['observation_variable']
    if len(referred_observations) == 0:
        break
    sample += list(referred_observations)

```
