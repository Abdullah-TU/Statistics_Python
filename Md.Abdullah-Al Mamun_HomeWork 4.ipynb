## 1.	Suppose we want to test whether the mean weight of apples in a grocery store is 150 grams. We randomly sample 20 apples from the store and measure their weights, getting the following data:

### Apple_weights = [145, 155, 160, 146, 142, 152, 150, 147, 148, 149, 148, 152, 153, 155, 154, 148, 151, 147, 153, 146]

### ○	What test should we use and why?

#### We will do hypothesis testing for a single group with an external standard, where we want to test whether the mean weight of a sample of apples is significantly different from a hypothesized population mean of 150 grams. 

### We will do:
  ### one-sample t-tests
  - to test hypotheses about a population mean based on a single sample of data.


## We use a t-test because we don't know the population standard deviation, and we are estimating it from the sample.


# ○ State the null and alternative hypotheses.

#### Null hypothesis: 
#### H0: μ = 150

#### Alternative hypothesis: 

#### H1: μ ≠ 150

# ○ Choose a significance level (α) (the probability of rejecting the null hypothesis when it is actually true).

```python
# library

import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### data

```python
Apple_weights = [145, 155, 160, 146, 142, 152, 150, 147, 148, 149, 148, 152, 153, 155, 154, 148, 151, 147, 153, 146]

Apple_weights
```

```python
μ = 150
```

```python
random_sample= random.sample(Apple_weights, k=20)
random_sample
```

```python
sample_mean = np.mean(random_sample)
print("The mean of the sample: {:.2f}".format(sample_mean))

sample_std = np.std(random_sample, ddof=1)
print("The standard deviation of the sample: {:.2f}".format(sample_std))
```

```python
sample_size= len(random_sample)
print(sample_size)
```

## Determine the degrees of freedom (df) of the sample. Df = sample size -1

```python
Df = sample_size - 1
Df
```

## ○ Determine the critical value of t based on the significance level and degrees of freedom. For a two-tailed test with α = 0.05 and df = 19, the critical values are -2.093 and 2.093.

#### For a two-tailed test, we need to find the t-value that cuts off 2.5% of the probability in each tail of the distribution.

```python
α =0.05
```

```python
Df
```

```python
import scipy.stats as stats

lower_critical_value = stats.t.ppf(α/2, Df)
upper_critical_value = stats.t.ppf(1 - α/2, Df)

print("For a two-tailed test with α = 0.05 and df = 19, the critical values are {:.3f} and {:.3f}.".format(lower_critical_value,upper_critical_value))
```

## ○ Compare and interpret the results of the test to the critical value

```python
import math

# t value
t_value = (sample_mean - μ) / (sample_std / math.sqrt(sample_size))
print("t-value = {:.3f}".format(t_value))

if lower_critical_value < t_value < upper_critical_value:
    print("Fail to reject the null hypothesis. There is not enough evidence to suggest a difference between the mean weight of apples in the grocery store and the hypothesized value of 150 grams.")
else:
    print("Reject the null hypothesis. There is enough evidence to suggest a difference between the mean weight of apples in the grocery store and the hypothesized value of 150 grams.")
```
